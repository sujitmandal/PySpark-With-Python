```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('test 3').getOrCreate()
```

```python
df = spark.read.csv('dataset/test2.csv',header=True,inferSchema=True)
df.show(3)
```

```python
df.filter("Item_MRP<=100").show()
```

```python
df.filter("Item_MRP<=200").show()
```

```python
df.filter("Item_MRP<=500").select('Item_Identifier', 'Item_MRP').show()
```

```python
df.filter(df['Item_MRP']<=500).show()
```

```python
df.filter((df['Item_MRP']<=500) | (df['Item_MRP']>=1000)).show()
```

```python
df.filter((df['Item_MRP']>=500) | (df['Item_MRP']<=1000)).show()
```

```python
df.filter(~(df['Item_MRP']<=50)).show()
```

```python
df.filter(df['Item_MRP'].between(100, 500)).show()
```

```python
df.filter(df['Item_MRP'].between(50, 100)).show()
```

```python
df2 = df.withColumnRenamed('Item_Identifier', 'Item')
df2 = df2.withColumnRenamed('Item_Weight', 'Weight')
df2 = df2.withColumnRenamed('Item_MRP', 'MRP')
df2 = df2.withColumnRenamed('Item_MRP_imputed', 'MRP_Imputed')
df2.show()
```

```python
df3 = df2.drop('Item_Weight + 2')
df3 = df3.drop('Item_Weight_imputed')
df3.show()
```

```python
df3.toPandas().to_csv("dataset/test3.csv", header=True, index=False)
```
